```python
import numpy as np
import pandas as pd
import surfboard.feature_extraction as surf
import os
from statsmodels.stats.anova import AnovaRM
```

## Ekstracja cech, parametrów

#### ateraz1 dziala

```python
files = []
for r,d,f in os.walk('audio'):
    for file in f:
        if 'wav' in file:
            files.append('audio/' + file)
        
files
```

```python
len(files)
```

```python
waves = surf.load_waveforms_from_paths(paths=files, sample_rate=48000)
```

```python
res = surf.extract_features(
    waves, 
    components_list=['rms', 'spectral_entropy', 'f0_statistics', 'hnr', 'f0_contour'],
    statistics_list=['mean', 'std', 'first_quartile', 'second_quartile', 'third_quartile', 'max', 'min']
)
```

```python
def get_subject(filepath):
    subject = filepath[6]
    return subject
    
def get_condition(filepath):
    # cut unitl the end
    cond = filepath[8:]
     #cut extension
    cond = cond[:-4]
    return cond
    

res['filename'] = files
res['subject'] = res.filename.apply(get_subject)
res['condition'] = res.filename.apply(get_condition)
```

## Analiza statystyczna

```python
df_pos = df[(df["condition"] == "pos")]
df_neg = df[(df["condition"] == "neg")]
df_neu = df[(df["condition"] == "neu")]
```

```python
import scipy.stats as stata
```

```python
#pos
rms_array_pos = df_pos[['rms_mean']].to_numpy()
f0_array_pos = df_pos[['f0_mean']].to_numpy()
hnr_array_pos = df_pos[['hnr']].to_numpy()

#neg
rms_array_neg = df_neg[['rms_mean']].to_numpy()
f0_array_neg = df_neg[['f0_mean']].to_numpy()
hnr_array_neg = df_neg[['hnr']].to_numpy()

#neu
rms_array_neu = df_neu[['rms_mean']].to_numpy()
f0_array_neu = df_neu[['f0_mean']].to_numpy()
hnr_array_neu = df_neu[['hnr']].to_numpy()
```

```python
print("pos","\n"
      "shapir_hnr:", stata.shapiro(hnr_array_pos),"\n"
      "shapir_f0:", stata.shapiro(f0_array_pos),"\n"
      "shapir_rms:", stata.shapiro(rms_array_pos),"\n")

print("neg","\n"
      "shapir_hnr:", stata.shapiro(hnr_array_neg),"\n"
      "shapir_f0:", stata.shapiro(f0_array_neg),"\n"
      "shapir_rms:", stata.shapiro(rms_array_neg),"\n")

print("neu","\n"
      "shapir_hnr:", stata.shapiro(hnr_array_neu),"\n"
      "shapir_f0:", stata.shapiro(f0_array_neu),"\n"
      "shapir_rms:", stata.shapiro(rms_array_neu))
```

```python
# feature - RMS
aovrm2way_1 = AnovaRM(df, depvar='rms_mean', subject='subject', within=['condition'], aggregate_func="mean" )
df_rms = aovrm2way_1.fit()
print("RMS", df_rms)

# feature - f0
aovrm2way_2 = AnovaRM(df, depvar='f0_mean', subject='subject', within=['condition'] )
df_f0 = aovrm2way_2.fit()
print("f0", df_f0)

# feature - hnr
aovrm2way_3 = AnovaRM(df, depvar='hnr', subject='subject', within=['condition'] )
df_hnr = aovrm2way_3.fit()
print("hnr", df_hnr)
```
